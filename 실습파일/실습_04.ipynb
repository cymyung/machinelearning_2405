## matplotlib.pyplot

```python
### 필요한 라이브러리.모듈 임폴트
import matplotlib.pyplot as plt
```

```python
### 그래프 생성 --> 환경 설정 --> rcParams
print(f'기본 설정값 : \n{plt.rcParams}')
```

```python
### 환경 변수 --> 그래프의 크기(기본값) --> 재설정
plt.rcParams['figure.figsize'] = [4,3]

# 결과 확인
print(plt.rcParams['figure.figsize'])
```

### 직선 그래프

```python
# 데이터 생성
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 8, 27, 64]

# 그래프 생성 전 --> 그래프의 크기 재설정
plt.figure(figsize=(6.4,4.8))

# 그래프 생성 --> plt.show(X) --> 두 그래프 중첩
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()

print('-'*80)

# 그래프 생성 전 --> 그래프의 크기 재설정
plt.figure(figsize=(6,6))

# 두 그래프 생성 --> 각각 그래프 출력
plt.plot(x, y1)
plt.show()

print('-'*80)

plt.plot(x, y2)
plt.show()

# # 그래프 출력
# plt.show()
```

```python
### 직선의 색깔 지정

# 데이터 --> x, y1 사용

# plot(color='색') --> 그래프 생성
plt.plot(x, y1, color='green')
# plt.plot(x, y1, color='#008000')
plt.show()
```

```python
### 직선의 모양 지정

# 데이터 생성 --> x, y1

# plt.plot(x, y1, linestyle='') --> 그래프 생성
# plt.plot(x, y1, linestyle='-.')
plt.plot(x, y1, linestyle=':')
plt.show()
```

```python
### 축 레이블 설정하기

# 데이터 생성 --> x, y1

# 직선 그래프 생성
plt.plot(x, y1)

# 축 레이블 설정
plt.xlabel("X axis")
plt.ylabel("Y axis")

# 그래프 출력
plt.show()
```

```python
### title 설정하기

# 데이터 생성 --> x, y1

# 그래프 생성
plt.plot(x, y1)

# 제목 설정
plt.title("Sample graph")

# 그래프 저장 --> plt.savefig(file_path) 실행
file_path='/content/drive/MyDrive/데이터분석/sample.png'
plt.savefig(fname=file_path)

# 그래프 출력
plt.show()
```

### 막대 그래프

```python
### 연도별 직원수 시각화

# 데이터 생성
year = ['2017','2018','2019']
# year = [2017, 2018, 2019]
number = [10, 40, 90]

# 막대 그래프 생성
plt.bar(x=year, height=number, color=['orange','green','blue'])

# 축 레이블 설정
plt.xlabel("year")
plt.ylabel("number")

# 제목 설정
plt.title("number of employees")

# 그래프 출력
plt.show()
```

### 히스토그램

```python
### 체중별 빈도수 시각화

# 데이터 생성
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

# 그래프 생성 및 출력 --> 기본 설정 : 구간화 적용(bins=10)
plt.hist(x=weight)
plt.show()

print('-'*80)

# 그래프 생성 및 출력 --> bins=100
plt.hist(x=weight, bins=100)
plt.show()
```

## seaborn

```python
### 필요한 라이브러리 임폴트
import seaborn as sns
```

```python
### 데이터 생성 --> tips 데이터

'''
1. seaborn 제공하는 데이터 : sns.load_dataset('tips')
2. tips 컬럼 설명
total_bill: 총 합계 요금표
tip: 팁
sex: 성별
smoker: 흡연자 여부
day: 요일
time: 식사 시간
size: 식사 인원
'''

# 데이터 다운로드
df_tips = sns.load_dataset('tips')

# 결과 확인하기
print(df_tips)
```

### 직선 그래프

```python
### 요금 총합과 팀의 관계 시각화

# 그래프 생성 --> lineplot()
sns.lineplot(data=df_tips, x='total_bill', y='tip')
plt.show()

print('-'*80)

# hue --> 추가 분석 자동화, 'time'적용
sns.lineplot(data=df_tips, x='total_bill', y='tip', hue='time')
plt.show()
```
