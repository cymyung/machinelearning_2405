## matplotlib.pyplot

```python
# pyplot impot
import matplotlib.pyplot as plt

# google에서 pip 검색
# pip에서 matplotlib이 등록되어 있는지 검색
# matplotlib 에서 github로 어떻게 들어갔지?
```

```python
### 그래프 생성 --> 환경설정 --> rcParams 이라는 변수 안에 저장되어 있다.
print(f'기본설정값 : \n{plt.rcParams}')
# figure.figsize: [6.4,4.8]   등에서 기본설정이 이미 되어 있다. -> 수치 단위는 인치다.
# font.family : ['sans-serif']  -> 이건 한글이 아니라 한글이 깨진다는 의미
```

```python
# 만약 환경설정에서 그래프 생성전 크기 6.4, 4.8 크기를 변경하고 싶다면?
### 환경변수 --> 그래프의 크기(기본값) --> 재설정

plt.rcParams['figure.figsize'] = [4,3]
print(plt.rcParams['figure.figsize'])
```

```python
# plot : plot y verses x as lines and or markers : line을 그려주는 함수
# parameter : 매개변수 , arguments : 인수
# *args : list 형태로 값을 주라는 뜻
```

### 직선그래프

```python
# 데이터생성
x = [1,2,3,4]
y1 = [1,4,9,16]
y2 = [1,8,27,64]

# 그래프 생성전에  --> 그래프의 크기 설정
plt.figure(figsize = (3,2))

# 그래프생성
plt.plot(x,y1)
plt.plot(x,y2)

# 그래프 출력
plt.show()

# -> 이건 두개의 그래프가 겹치는 거고..
# 크기는 그래프 생성전에 바꿔야 한다.
```

```python
# 이건 각각의 그래프가 출력 되는 형태
# 그래프 생성 후 그래프 출력
# plt.show()는 그래프 생성 후 이를 마무리하는 의미.
plt.figure(figsize=(10,2))
plt.plot(x,y1, color = 'magenta')
plt.show()

# 그래프 생성 후 그래프 출력
# 그래프 선의 색깔

plt.plot(x,y2, color = 'y')
plt.show()
```

```python
### 직선의 색깔 지정

# 데이터 --> x, y1 사용
# plot(color = '색')  --> 그래프 생성
plt.plot(x,y1, color = '#008000', linewidth = 3, linestyle = '-.')
plt.show()
```

```python
###  직선의 모양 지정
# 데이터 생성 -> x, y1
# plt.plot(x,y1, linestyle = '-')
# '-', '-.', ':'
plt.plot(x,y1,linestyle = ':')
plt.show()
```

```python
# 축 레이블 설정하기
# 데이터 생성 --> x, y1

# 직선그래프 생성
plt.plot(x,y1)

# 축 레이블 설정
plt.xlabel('X axis')
plt.ylabel('Y axis')

# 타이틀 설정하기
plt.title('sample graph')

# 그래프 출력
plt.show()
```

```python
# 그래프 저장하기
# 1. 마우스 오른쪽 버튼 '새이름으로 저장'
# 2. matplotlib  에서 그래프 저장하기 기능 있음.
#    -> matplotlib.pyplot.savefig

plt.plot(x,y1)

# 축 레이블 설정
plt.xlabel('X axis')
plt.ylabel('Y axis')

# 타이틀 설정하기
plt.title('sample graph')

# plt.savefig(file_path) 실행
file_path = '/content/drive/MyDrive/데이터분석/sample.png'
plt.savefig(fname = file_path)
```

## searborn

```python
# seaborn은 pandas dataframe을 기본으로 한다.
# matplotlib과는 상황이 다르다.

import seaborn as sns
```

```python
# seaborn 은 dataframe에 적합하므로, df를 만들고 시작

### 데이터 생성 --> tips 데이터

'''
1. seaborn 제공하는 데이터 : sns.load_dataset('tips')
2. tips 컬럼 설명
total_bill: 총 합계 요금표
tip: 팁
sex: 성별
smoker: 흡연자 여부
day: 요일
time: 식사 시간
size: 식사 인원
'''

# 데이터 다운로드
df_tips = sns.load_dataset('tips')

# 결과 확인하기
print(df_tips)
```

### 직선그래프

```python
### 요금 총합과 팁의 관계 시각화
# 그래프 생성 --> lineplot()

sns.lineplot(data = df_tips, x = 'total_bill', y = 'tip') # seaborn은 x, y label을 자동으로 넣어준다.
plt.show()

print('-'*80)

# hue를 이용 추가분석, 'time'식사시간을 적용하여 분리 :  이 hue가 seaborn의 뛰어난 기능
# -> 통계처리를 seaborn이 알아서 해준 후 시각화 해줘서 뛰어난 기능!!
sns.lineplot(data = df_tips, x = 'total_bill', y = 'tip', hue='time') # seaborn은 x, y label을 자동으로 넣어준다.
plt.show()
```

```python
# 원래는 이렇게 했어야 함.
# 보충 : hue기능을 사용하지 않고, 식사 시간대별로 분할해서 요금의 총합과 팁의 관계를 시각화 하는 경우

# 전체 데이터 --> 식사 시간대별로 분할
condition1 = df_tips.loc[:,'time'] == 'Lunch'
condition2 = df_tips.loc[:,'time'] == 'Dinner'

df_lunch  = df_tips.loc[condition1, :]
df_dinner = df_tips.loc[condition2, :]

print(df_lunch)
print('*'*80)
print(df_dinner)

print('*'*80)

sns.lineplot(data = df_lunch,  x = 'total_bill', y = 'tip')
plt.show()
sns.lineplot(data = df_dinner, x = 'total_bill', y = 'tip')
plt.show()

print('*'*80)

# 시각화
sns.lineplot(data = df_lunch,  x = 'total_bill', y = 'tip')
sns.lineplot(data = df_dinner, x = 'total_bill', y = 'tip')
plt.show()
#  이걸 seaborn은 hue를 이용해서 간단하게 처리해줌.
```

```python
sns.scatterplot(data = df_tips, x = 'total_bill', y = 'tip')
plt.show()
```

### 막대그래프

```python
# seaborn에서는 countplot 으로 하는데, 이 하나로 수평, 수직 막대그래프를 전부 이용
# pyplot 은 bar, barh 두개 별도의 함수 존재

### 시각화 목표 : 요일별 빈도수

# x 축에 수직인 막대그래프 생성
# seaborn 의 색깔세트를 이용하려면 pallete 매개변수를 이용할 수 있다.
# seaborn  홈페이지에서 확인

sns.countplot(data = df_tips, x = 'day', palette = 'husl')
plt.show()

# y 축에 수직인 막대그래프 생성
sns.countplot(data = df_tips, y = 'day', palette = 'Set2')
plt.show()
```

```python
# 위에서 + 손님의 성별을 추가

# x 축에 수직인 막대그래프 생성
sns.countplot(data = df_tips, x = 'day', hue = 'sex', palette = 'flare')
plt.show()

# y 축에 수직인 막대그래프 생성
sns.countplot(data = df_tips, y = 'day', hue = 'sex', palette = 'Spectral')
plt.show()
```

### 산점도 그래프

```python
### 시각화 목표 : 요금총합과 팁의 관계(상관관계)

# 그래프 생성 및 출력
sns.scatterplot(data = df_tips, x = 'total_bill', y = 'tip', hue = 'sex')
plt.show()

print('-'*80)
# 추가분석 : 상관계수 --> df.corr()
df_tips.loc[:,['total_bill','tip']].corr()
```

```python
# 추가분석 : 전체 데이터 --> 식사시간대로 분할

condition1 = df_tips.loc[:,'time'] == 'Lunch'
condition2 = df_tips.loc[:,'time'] == 'Dinner'

df_lunch  = df_tips.loc[condition1, :]
df_dinner = df_tips.loc[condition2, :]

lunch_corr  = df_lunch.loc [:, ['total_bill','tip']].corr()
dinner_corr = df_dinner.loc[:, ['total_bill','tip']].corr()

print(f"점심시간대의 요금 총합과 팁의 상관계수 : \n{lunch_corr}")
print('-'*80)
print(f"저녁시간대의 요금 총합과 팁의 상관계수 : \n{dinner_corr}")
```

###heatmap

```python
# heatmap의 data는 상관행렬을 넣는다.
# data : 상관행렬
# annot : 상관계수 표시 여부
# annot(annotation)은 그래프에 숫자가 표시되도록.. 한다.(True, False)
# fmt : 상관계수의 소수점 자리 수를 지정 예시 : .2f
# linewidth : cell 사이의 간격을 조정

# 상관행렬을 구하기 위한 상관분석 시행
tip_corr = df_tips.corr(numeric_only = True)
print(tip_corr)

sns.heatmap(data = tip_corr, annot = True, fmt = '.2f', cmap = 'gray')
```
