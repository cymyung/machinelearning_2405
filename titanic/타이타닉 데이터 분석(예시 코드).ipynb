# 타이타닉 데이터 분석

## 데이터 불러오기

```python
# 필요한 라이브러리 임폴트
import pandas as pd
```

```python
# 파일 경로 설정
train_path = '/content/drive/MyDrive/데이터분석/titanic_train.csv'
test_path = '/content/drive/MyDrive/데이터분석/titanic_test.csv'

# pd.read_csv() 사용 --> DataFrame 생성
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# 결과 확인하기
print(df_train)
print('-'*80)
print(df_test)
```

```python
# df.info() --> 행과 컬럼에 대한 요약 정보를 출력하는 함수
df_train.info()
print('-'*80)
df_test.info()
```

```python
### 데이터 병합하기
df = pd.concat([df_train, df_test], ignore_index=True)
print(df)
print('-'*80)
df.info()
```

## 데이터 전처리

### 누락 데이터 처리

```python
### 각 컬럼별 누락 데이터의 수 확인

# isnull().sum() 사용
num_nulls = df.isnull().sum()
print(f'각 컬럼별 누락 데이터의 수 : \n{num_nulls}')
```

#### Cabin 컬럼

```python
### Cabin 컬럼 제거 --> df.drop() 함수 사용
cleaned_df = df.drop(columns=['Cabin'])
print(f'Cabin 컬럼을 제거한 결과 : \n{cleaned_df}')
```

#### Age 컬럼

```python
### Age 컬럼 값의 분포 --> 요약 통계량 추출
print(cleaned_df.loc[:, 'Age'].describe())
```

```python
### Age 컬럼 --> 나이별 빈도수 --> 최빈값 확인
counts = cleaned_df.loc[:, 'Age'].value_counts()
print(f'Age 컬럼의 나이별 승객 수 : \n{counts}')
print('-'*80)
print(f'Age 컬럼의 나이별 승객의 수 top10 : \n{counts.iloc[0:10]}')
print('-'*80)
print(f'Age 컬럼의 나이별 승객의 수 top1 : \n{counts.index[0]}')
```

```python
### Age 컬럼 --> 누락 데이터 --> 최빈 값으로 대체
cleaned_df.loc[:, 'Age'].fillna(counts.index[0], inplace=True)

# 최빈 값으로 대체 후 누락 데이터 수 확인
num_nulls = cleaned_df.loc[:, 'Age'].isnull().sum()
print(f'Age 컬럼의 누락 데이터의 수 : {num_nulls}')
```

##### Binning

```python
### Age 컬럼 --> 8단계 구간화

# 매개 변수에 전달할 값(arguments) 설정
x = cleaned_df.loc[:, 'Age']
bins = 8
labels = [0, 1, 2, 3, 4, 5, 6, 7]

# pd.cut() 함수 사용
cleaned_df.loc[:, 'Age'] = pd.cut(x=x, bins=bins, labels=labels)

# 결과 확인하기
print(f'데이터프레임 확인 : \n{cleaned_df}')
```

#### Embarked 컬럼

```python
### 승선 항구의 항구별 빈도수 / 비율 추출

# 빈도수 추출
counts = cleaned_df.loc[:, 'Embarked'].value_counts()
print(f'승선 항구별 빈도수 : \n{counts}')

print('-'*80)

# 비율 추출
ratio = cleaned_df.loc[:, 'Embarked'].value_counts(normalize=True)
print(f'승선 항구별 비율 : \n{ratio}')
```

```python
### 최빈 값으로 채우기
cleaned_df.loc[:, 'Embarked'].fillna('S', inplace=True)

# 결과 확인하기 --> 누락 데이터의 수 확인
num_nulls = cleaned_df.loc[:, 'Embarked'].isnull().sum()
print(f'Embarked 컬럼의 누락 데이터의 수 : {num_nulls}')
```

#### Fare 컬럼

```python
### Fare 컬럼의 항목별 빈도수 확인
counts = cleaned_df.loc[:, 'Fare'].value_counts()
print(f'요금의 금액별 빈도수 : \n{counts}')
```

```python
### Fare 컬럼에서 누락 데이터 확인 --> 1명 who?

# Fare 컬럼 --> 누락 여부 --> isnull() --> 누락인 행 True, 나머지 False
condition = cleaned_df.loc[:, 'Fare'].isnull()
print(f'Fare 컬럼에서 누락 여부를 확인한 결과 : \n{condition}')

print('-'*80)

# 블리언 인덱싱 --> df.loc[row, :]
data_null = cleaned_df.loc[condition, :]
print(f'Fare 컬럼에서 누락된 데이터 확인 : \n{data_null}')
```

```python
### 3등석 & 성별 'male' & 나이 구간 6 & 승선 항구 S & SibSp 0 & Parch 0 --> 승객의 요금 확인

# 조건식 생성
condition1 = (cleaned_df.loc[:, 'Pclass']==3)
condition2 = (cleaned_df.loc[:, 'Gender'] == 'male')
condition3 = (cleaned_df.loc[:, 'Age'] == 6)
condition4 = (cleaned_df.loc[:, 'Embarked'] == 'S')
condition5 = (cleaned_df.loc[:, 'SibSp'] == 0)
condition6 = (cleaned_df.loc[:, 'Parch']==0)

condition = condition1&condition2&condition3&condition4&condition5&condition6

# 6가지 조건을 만족하는 행 --> Fare 추출
fare = cleaned_df.loc[condition, 'Fare']
print(f'6가지 조건을 모두 만족하는 승객의 요금 : \n{fare}')
```

```python
### 누락 데이터 대체
price = fare.iloc[0]
cleaned_df.loc[:, 'Fare'].fillna(price, inplace=True)

# 결과 확인하기
print(cleaned_df.iloc[data_null.index, :])
```

### Feature Engineering

#### 가족 인원수 컬럼 추가

```python
### SibSp 컬럼 값과 Parch 컬럼 값 --> 덧셈 -->  가족 인원수 컬럼 생성 --> 파생 변수 생성

# SibSp 컬럼 값과 Parch 컬럼 값 --> 덧셈
sum = cleaned_df.loc[:, 'SibSp'] + cleaned_df.loc[:, 'Parch']
print(f'SipSp 컬럼과 Parch 컬럼의 덧셈 결과 : \n{sum}')

print('-'*80)

# 가족 인원수 컬럼 생성 --> 값 입력
cleaned_df.loc[:, 'num_family'] = sum

# 결과 확인하기
print(cleaned_df)
```

#### Name 컬럼 세분화

```python
### Name 컬럼 확인하기

"""
### 이름의 구성 : 성, 호칭. 이름
"""

# Name 컬럼의 값 확인
print(cleaned_df.loc[:, 'Name'])
```

```python
### '호칭' 추출 --> split() 함수

"""
### split() 함수
(예시)
name = 'Braund, Mr. Owen Harris'
name.split()
"""

# test name 생성
test_name = cleaned_df.loc[:, 'Name'].iloc[0]
print(test_name)

print('-'*80)

# 첫번째 분할 --> 기준 : ','
result1 = test_name.split(',')
print(f'첫번째 분할의 결과 : {result1}')

print('-'*80)

# 두번째 분할 --> 기준 : '.'
# result2 = result1[1].split('.')
result2 = test_name.split(',')[1].split('.')
print(f'두번째 분할의 결과 : {result2}')

print('-'*80)

# '호칭' 추출
# title = result2[0]
title = test_name.split(',')[1].split('.')[0]
print(f'추출한 호칭 확인 : {title}')
```

```python
### 문자열의 앞/뒤 공백 제거 --> strip() 함수

# 예시
string = '   Hello world, Python!'
print(f'strip 함수 사용 전 문자열 확인 : {string}')
print('-'*80)
result = string.strip()
print(f'strip 함수 사용 후 문자열 확인 : {result}')

print('-'*80)

# 추출된 호칭에 strip 함수 적용
# title1 = title.strip()
title1 = test_name.split(',')[1].split('.')[0].strip()
print(title1)
```

```python
### Name 컬럼 --> 호칭 추출 함수 정의
def extract_title(name):
    # ','로 split --> 두번째 성분 원소 선택 --> '.'로 split --> 첫번째 성분 원소 선택 --> strip 함수
    title = name.split(',')[1].split('.')[0].strip()
    return title
```

```python
### Name 컬럼으로부터 호칭을 추출 --> 새로운 컬럼 --> 파생 변수 생성

'''
# 호칭 정리
  1) 'Rev', 'Col', 'Major', 'Dr', 'Capt', 'Sir' : 직위 표현
  2) 'Ms', 'Mme', 'Mrs', 'Dona' : 여성 표현
  3) 'Miss', 'Mlle', 'Lady' : 젊은 여성 표현
  4) 'Mr', 'Don' : 남성 표현
  5) 'Master' : 주로 청소년 이하 결혼하지 않은 남성
  6) 'Jonkheer', 'the Countess' : 귀족 표현
'''

# Name 컬럼 --> extract_title 적용 --> 호칭 추출 --> 새로운 컬럼 생성
title = cleaned_df.loc[:, 'Name'].apply(extract_title)
print(f'호칭 추출의 결과 : \n{title}')

print('-'*80)

# title 컬럼 생성 + 추출된 호칭 입력
cleaned_df.loc[:, 'title'] = title

# 결과 확인하기
print(f'title 컬럼 추가 후 데이터프레임 확인 : \n:{cleaned_df}')
```

#### title 컬럼 정리

```python
### title 컬럼의 항목별 빈도수 / 비율 확인

# 빈도수 추출
counts = cleaned_df.loc[:, 'title'].value_counts()
print(f'title 컬럼의 호칭별 빈도수 : \n{counts}')

print('-'*80)

# 비율 추출
ratio = cleaned_df.loc[:, 'title'].value_counts(normalize=True)
print(f'title 컬럼의 호칭별 비율 : \n{ratio}')
```

```python
### Series --> index 확인 --> '호칭' 추출
row_index = counts.index
print(f'행 인덱스 확인 : \n{row_index}')
```

```python
### title 컬럼에서 Mr 또는 Miss 또는 Mrs 또는 Master가 아닌 값 --> other로 변환(1)

# 데이터프레임 복사
df_test = cleaned_df.copy()
# df_test = cleaned_df.iloc[:, :]

# replace() 함수 사용 --> 호칭 변환
df_test.loc[:, 'title'] = df_test.loc[:, "title"].replace({row_index[4]:'other',
                                                           row_index[5]:'other',
                                                           row_index[6]:'other',
                                                           row_index[7]:'other',
                                                           row_index[8]:'other',
                                                           row_index[9]:'other',
                                                           row_index[10]:'other',
                                                           row_index[11]:'other',
                                                           row_index[12]:'other',
                                                           row_index[13]:'other',
                                                           row_index[14]:'other',
                                                           row_index[15]:'other',
                                                           row_index[16]:'other',
                                                           row_index[17]:'other'}
                                                          )

# 결과 확인하기 : df_test --> title 컬럼 --> 호칭별 빈도수
counts = df_test.loc[:, 'title'].value_counts()
print(f'호칭 변환 후 title 컬럼의 호칭별 빈도수 : \n{counts}')
```

```python
### title 컬럼에서 Mr 또는 Miss 또는 Mrs 또는 Master가 아닌 값 --> other로 변환(2)

# 비교 연산자를 이용 --> 블리언 배열 생성
condition1 = (cleaned_df.loc[:, 'title'] != 'Mr')
condition2 = (cleaned_df.loc[:, 'title'] != 'Mrs')
condition3 = (cleaned_df.loc[:, 'title'] != 'Miss')
condition4 = (cleaned_df.loc[:, 'title'] != 'Master')

condition = condition1 & condition2 & condition3 & condition4
print(f'4가지 조건을 모두 만족하는 경우 : \n{condition}')

print('-'*80)

# 비교 연산자의 결과 --> 블리언 인덱싱 --> other로 수정
cleaned_df.loc[condition, 'title'] = 'other'

# 결과 확인하기 --> title 컬럼의 항목별 빈도수 추출
counts = cleaned_df.loc[:, 'title'].value_counts()
print(f'title 컬럼의 항목별 빈도수 : \n{counts}')
```
