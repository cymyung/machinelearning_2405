# 누락 데이터 처리하기

## 누락된 데이터의 수 확인

```python
### 각 컬럼별 누락 데이터의 수 확인

# isnull() 사용 --> 각 컬럼의 모든 데이터 --> 누락 여부 확인 --> 블리언 값 표시
print(df.isnull())

print('-'*80)

# isnull().sum() 사용 --> 각 컬럼별 누락 데이터의 수 확인
num_nulls = df.isnull().sum()
print(f'각 컬럼별 누락 데이터의 수 : \n{num_nulls}')
```

```python
### 특정 컬럼 --> 누락 데이터가 존재하는 행 index 확인

## 컬럼 --> 도서명 컬럼

# 비교 연산의 기능 --> isnull() 적용
condition = df.loc[:, '도서명'].isnull()
print(f'isnull() 함수를 적용한 결과 : \n{condition}')

print('-'*80)

# 비교 연산의 결과 --> 블리언 인덱싱
index_nulls = df.loc[:, '도서명'].loc[condition].index
print(f'도서명 컬럼에서 누락 데이터의 행 인덱스 추출 : \n{index_nulls}')
```

## 누락 데이터 제거

```python
### df.dropna() --> 누락된 데이터가 존재하는 모든 행 삭제
df2 = df.dropna(ignore_index=True)
print(f'누락된 데이터가 존재하는 모든 행 삭제의 결과 : \n{df2}')
```

```python
### 도서명 컬럼에 존재하는 누락 데이터 제거

"""
1. 도서명 컬럼에 존재하는 누락 데이터의 행 인덱스 이용
2. 사용 함수 : df.drop(index, columns)
"""

df3 = df.drop(index=index_nulls)

# 결과 확인하기
print(f'전체 데이터에서 도서명 컬럼에 존재하는 누락 데이터를 제거한 결과 : \n{df3}')
```

```python
### 보충 : drop() --> 특정 데이터를 제거

# 출판사 이름이 '한빛미디어'인 데이터만 제거
condition = (df.loc[:, '출판사'] == '한빛미디어')
target = df.loc[condition, :].index
print(f'출판사 이름이 한빛미디어인 행의 인덱스 : \n{target}')

print('-'*80)

# drop(index) 사용
print(df.drop(index=target))
```

```python
### 도서명 컬럼과 저자 컬럼에 존재하는 누락 데이터 제거

"""
1. dropna() 사용
2. 매개변수 --> subset 이용
"""

df4 = df.dropna(subset=['도서명', '저자'], ignore_index=True)

# 결과 확인하기
print(f'전체 데이터에서 도서명 컬럼과 저자 컬럼에 존재하는 누락 데이터를 제거한 결과 : \n{df4}')
```

# 이상치 처리하기

```python
### 데이터 불러오기

# 필요한 라이브러리 임폴트
import pandas as pd

# 파일 경로 설정
file_path = '/content/drive/MyDrive/데이터분석/weather.csv'

# pd.read_csv() --> DataFrame 생성
df_weather = pd.read_csv(file_path)

# 결과 확인하기
print(df_weather)
```

```python
### 요약 통계량 추출

# 각 컬럼별 8가지 통계량 추출
df_stat = df_weather.describe()

# 결과 확인하기
print(df_stat)
```

```python
### 연습 : 온도 컬럼 --> 1사분위 값, 3사분위 값 추출

# 1사분위 값 --> 인덱싱
# q1 = df_stat.loc['25%', 'temp']

# 1사분위 값 --> df.col.quantile(q=0.25)
q1 = df_weather.loc[:, 'temp'].quantile(0.25)
print(f'온도 컬럼의 1사분위 값 : {q1}')

print('-'*80)

# 3사분위 값 --> 인덱싱
# q3 = df_stat.loc['75%','temp']

# 3사분위 값 --> df.col.quantile(q=0.75)
q3 = df_weather.loc[:, 'temp'].quantile(0.75)
print(f'온도 컬럼의 3사분위 값 : {q3}')
```

## 온도 컬럼의 이상치 제거

```python
### 온도 컬럼의 첫번째 이상치 추출

# 1사분위 값 / 3사분위 값
q1 = df_weather.loc[:, 'temp'].quantile(q=0.25)
q3 = df_weather.loc[:, 'temp'].quantile(q=0.75)
print(f'온도 컬럼의 첫번째 q1 : {q1}')
print('-'*80)
print(f'온도 컬럼의 첫번째 q3 : {q3}')

print('-'*80)

# 첫번째 iqr 측정
iqr = q3 - q1
print(f'온도 컬럼의 첫번째 iqr = {iqr}')

print('-'*80)

# 정상 범위의 첫번째 최소 / 최대 값 측정
min = q1 - (iqr*1.5)
max = q3 + (iqr*1.5)
print(f'온도 컬럼의 첫번째 정상 범위의 최소값 : {min}')
print('-'*80)
print(f'온도 컬럼의 첫번째 정상 범위의 최대값 : {max}')

print('-'*80)

# 비교 연산 + 논리 연산 적용 --> 조건 설정
condition = (df_weather.loc[:, 'temp'] < min) | (df_weather.loc[:, 'temp'] > max)
print(f'온도 컬럼의 첫번째 이상치에 대한 블리언 배열 : \n{condition}')

print('-'*80)

# 논리 연산의 결과 --> 블리언 인덱싱 --> 이싱치 데이터 index 추출
outlier_index = df_weather.loc[condition, 'temp'].index
print(f'온도 컬럼의 첫번째 이상치 데이터의 인덱스 : \n{outlier_index}')

print('-'*80)

# 온도 컬럼에 대한 첫번째 이상치 제거
if len(outlier_index) > 0:
    cleaned_temp1 = df_weather.loc[:, 'temp'].drop(index=outlier_index)
    print(f'온도 컬럼에 대한 첫번째 이상치 제거의 결과 : \n{cleaned_temp1}')
```

```python
### 온도 컬럼의 두번째 이상치 추출

# 1사분위 값 / 3사분위 값
q1 = cleaned_temp1.quantile(q=0.25)
q3 = cleaned_temp1.quantile(q=0.75)
print(f'온도 컬럼의 두번째 q1 : {q1}')
print('-'*80)
print(f'온도 컬럼의 두번째 q3 : {q3}')

print('-'*80)

# 두번째 iqr 측정
iqr = q3 - q1
print(f'온도 컬럼의 두번째 iqr = {iqr}')

print('-'*80)

# 정상 범위의 두번째 최소 / 최대 값 측정
min = q1 - (iqr*1.5)
max = q3 + (iqr*1.5)
print(f'온도 컬럼의 두번째 정상 범위의 최소값 : {min}')
print('-'*80)
print(f'온도 컬럼의 두번째 정상 범위의 최대값 : {max}')

print('-'*80)

# 비교 연산 + 논리 연산 적용 --> 조건 설정
condition = (cleaned_temp1 < min) | (cleaned_temp1 > max)
print(f'온도 컬럼의 두번째 이상치에 대한 블리언 배열 : \n{condition}')

print('-'*80)

# 논리 연산의 결과 --> 블리언 인덱싱 --> 이싱치 데이터 index 추출
outlier_index = cleaned_temp1.loc[condition].index
print(f'온도 컬럼의 두번째 이상치 데이터의 인덱스 : \n{outlier_index}')

print('-'*80)

# 온도 컬럼에 대한 두번째 이상치 제거
if len(outlier_index) > 0:
    cleaned_temp2 = cleaned_temp1.drop(index=outlier_index)
    print(f'온도 컬럼에 대한 두번째 이상치 제거의 결과 : \n{cleaned_temp2}')
```

## 풍속 컬럼의 이상치 제거

```python
### 풍속 컬럼의 첫번째 이상치 추출

# 1사분위 값 / 3사분위 값
q1 = df_weather.loc[:, 'speed'].quantile(q=0.25)
q3 = df_weather.loc[:, 'speed'].quantile(q=0.75)
print(f'풍속 컬럼의 첫번째 q1 : {q1}')
print('-'*80)
print(f'풍속 컬럼의 첫번째 q3 : {q3}')

print('-'*80)

# 첫번째 iqr 측정
iqr = q3 - q1
print(f'풍속 컬럼의 첫번째 iqr = {iqr}')

print('-'*80)

# 정상 범위의 첫번째 최소 / 최대 값 측정
min = q1 - (iqr*1.5)
max = q3 + (iqr*1.5)
print(f'풍속 컬럼의 첫번째 정상 범위의 최소값 : {min}')
print('-'*80)
print(f'풍속 컬럼의 첫번째 정상 범위의 최대값 : {max}')

print('-'*80)

# 비교 연산 + 논리 연산 적용 --> 조건 설정
condition = (df_weather.loc[:, 'speed'] < min) | (df_weather.loc[:, 'speed'] > max)
print(f'풍속 컬럼의 첫번째 이상치에 대한 블리언 배열 : \n{condition}')

print('-'*80)

# 논리 연산의 결과 --> 블리언 인덱싱 --> 이싱치 데이터 index 추출
outlier_index = df_weather.loc[condition, 'speed'].index
print(f'풍속 컬럼의 첫번째 이상치 데이터의 인덱스 : \n{outlier_index}')

print('-'*80)

# 풍속 컬럼에 대한 첫번째 이상치 제거
if len(outlier_index) > 0:
    cleaned_speed1 = df_weather.loc[:, 'speed'].drop(index=outlier_index)
    print(f'풍속 컬럼에 대한 첫번째 이상치 제거의 결과 : \n{cleaned_speed1}')
```
